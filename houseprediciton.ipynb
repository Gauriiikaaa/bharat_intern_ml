```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
```

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'condition']]
y = df['price']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)

# Make predictions
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# Code for loading data, model training etc
import matplotlib.pyplot as plt
# Make predictions
y_pred = reg.predict(X_test)

# Plot actual vs predicted
plt.scatter(y_test, y_pred)

plt.xlabel('Actual Price') 
plt.ylabel('Predicted Price')
plt.title('Linear Regression Price Prediction') 

# Plot perfect prediction line
xmin = min(y_test.min(), y_pred.min())
xmax = max(y_test.max(), y_pred.max())
plt.plot([xmin, xmax], [xmin, xmax], 'k--', lw=2)

plt.show()
```
