```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('wineQT.csv')
X = df.drop('quality', axis=1)
y = df['quality']
```

```python
poly = PolynomialFeatures(degree=2)
pipe = make_pipeline(poly, Lasso())
```

```python
X_poly = poly.fit_transform(X)
scores = cross_val_score(pipe, X_poly, y, scoring='neg_mean_squared_error', cv=5)
print("Lasso RMSE: ", np.mean(np.sqrt(-scores)))
```

```python
ridge = make_pipeline(poly, Ridge())
scores = cross_val_score(ridge, X_poly, y, scoring='neg_mean_squared_error', cv=5)
print("Ridge RMSE: ", np.mean(np.sqrt(-scores)))
```
